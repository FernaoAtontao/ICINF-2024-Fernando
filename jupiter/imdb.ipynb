```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataIMDB = pd.read_csv("data/datanueva.csv", index_col="ID")
dataIMDB.head()
```

<h4><b>DIMENSIÓN DEL DATASET</b></h4>

```python
print(f'Cantidad de Filas y Columnas del Dataset: {dataIMDB.shape}')
```

<h4><b>IDENTIFICANDO LAS COLUMNAS CON DATOS NULOS</b></h4>

```python
dataIMDB.isnull().any()
```

<h4><b>IDENTIFICANDO LAS FILAS EN ESPECIFICAS CON DATOS NULOS</b></h4>

```python
# Filtrar las filas que contienen valores nulos
filas_nulas = dataIMDB[dataIMDB.isnull().any(axis=1)]

filas_nulas
```

<h4><b>IDENTIFICANDO LAS FILAS EN ESPECIFICAS CON DATOS NULOS DE LA COLUMNA META SCORE</b></h4>

```python
filas_nulas_metascore = dataIMDB[dataIMDB['Meta_score'].isnull()]
filas_nulas_metascore
```

<h4><b>MOSTRAR LA CANTIDAD DE FILAS NULAS DE LAS COLUMNAS CERTIFICATE y METASCORE</b></h4>

```python
cant_f_nulascertificate = dataIMDB['Certificate'].isnull().sum()
cant_f_nulasscore = dataIMDB['Meta_score'].isnull().sum()
print(f'La cantidad de filas nulas de la columna Certificate es: {cant_f_nulascertificate}')
print(f'La cantidad de filas nulas de la columna Certificate es: {cant_f_nulasscore}')
```

<h4><b>ELIMINANDO NULOS DE LAS COLUMNAS CERTIFICATE y META SCORE</b></h4>

```python
df_sin_nulos = dataIMDB.dropna(subset=['Certificate', 'Meta_score'])
df_sin_nulos.count()
```

```python
df_sin_nulos.info()
```

```python
df_sin_nulos.head()
```

```python
promedio_por_categoria = df_sin_nulos.groupby('Director')['Meta_score'].mean()
promedio_por_categoria.head(10)
```

<h4><b>FUNCIÓN PARA VERIFICAR SI EL GENERO WAR ESTÁ PRESENTE EN LA LISTA Y GENERANDO UN NUEVO DATAFRAME</b></h4>

```python
def filtro_war(genres):
    return 'War' in genres

#Aplica el filtro utilizando apply
datosWAR = df_sin_nulos[df_sin_nulos['Genre'].apply(filtro_war)]

#Muestra el dataset solamente incluyendo la columnas Titulo y Ranting
peliswar = datosWAR[['Series_Title', 'IMDB_Rating', 'Meta_score']]
peliswar
```

<h4><b>GENERANDO UN GRÁFICO EN BARRA DE LAS PRIMERAS 10 PELICULAS DE GUERRA</b></h4>

<h4><b>COMPARANDO EL IMDB RATING CON EL META SCORE</b></h4>

```python
df_war = peliswar.head(10)
df_war.set_index('Series_Title', inplace=True)


ax = df_war.plot(kind='bar', figsize=(10, 6))

for i, rating in enumerate(df_war['IMDB_Rating']):
    ax.annotate(str(rating), xy=(i, rating), ha='center', va='bottom')

for i, score in enumerate(df_war['Meta_score']):
    ax.annotate(str(score), xy=(i, score), ha='center', va='bottom')

plt.show()
```
